```python
#Credits to staircase-scene: https://mitsuba.readthedocs.io/en/latest/src/gallery.html

import sys
import os 

mi_path_rel = r"../../mitsuba3/build/python"
mi_path = os.path.abspath(mi_path_rel)

sys.path.insert(0, mi_path)
import mitsuba as mi
```

```python
mi.set_variant("llvm_rgb")
```

```python
path_scene = os.path.abspath("./staircase/steady_scene.xml")
path_bitmap = os.path.abspath("./staircase/steady_scene.png")

assert os.path.exists(path_scene), f"Path of XML {path_scene} does not exist"

scene = mi.load_file(path_scene)
```

```python
integrator = scene.integrator()
sensors = scene.sensors()

data_steady = integrator.render(scene, sensor=sensors[0])
```

```python
mi.util.convert_to_bitmap(data_steady)
```

```python
bmp = mi.Bitmap(data_steady)

mi.util.write_bitmap(path_bitmap, data_steady, write_async=True)
```
